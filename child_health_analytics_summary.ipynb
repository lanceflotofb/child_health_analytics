## Introduction

The goal of this analysis was to determine the relationship, in children, between diet and being overweight or obese.

The data was gathered over the course of a few years by the center for disease control. The data was gathered on school children ages 3-15. Diet information was gathered from a one day sample in which the children were asked to keep track of what they ate (parents helped with the data collection). Their intake was then converted into the different nutrients that were in their foods.

In addition to the raw data, some additional features were created. Those features were macro deviation, nutrient deviation, and calorie excess. Calorie excess is exactly what it sounds like. It is the different between the amount of calories they should consume (based on weight, age, and gender) and the amount of calories they did consume. Macro deviation and nutrient deviation are measures of how close the student was to the FDA recommended daily values. The lower the number, the closer the child was to the FDA recommendation. FDA recommendations were adjusted for each child based on recommended calorie intake.

## Analysis

A variety of analysis was performed on the data. First the analysis was fit with a handful of machine learning models, and none of them fit the data very well. Given the number of feature vectors (64) and the minimal number of training samples (~850), I was concerned there was not enough datapoints to gather useful information from the features. For this reason, I performed feature elimination. I performed feature elimination with both a SVM and a linear model as these two have coefficents that can be used in recursive feature elimination. The features that were labeled as the most important were then given to the couple machine learning models that performed the best. Feature elimination did little to improve the fit of any of the models.

## Conclusions

Linear regression with feature elimination seems to be the most successful but it still does a poor job of fitting the data. 

The only variables with any meaning are height, age, and calorie excess. It seems that as height and calorie_excess increase, so does the likelihood of being overweight. As age increases, the likelihood of being overweight decreases. However, these three conclusions should be accepted with some skepticism. Although the P values were low for these specific variables, the R^2 and adjusted R^2 value suggested this model is very poorly fitted to the data as a whole.

More interesting however is what the data doesn't show to be important. It is very interesting that things like calories, carbs, fat, and macro_deviation, have seemingly no effect.

With all the data and analysis, I am lead to believe there are other factors not captured by the data, that are the true causal variables of being overweight or obese. Such factors could include things like exercise and play, genetic factors, and stress / happiness.
